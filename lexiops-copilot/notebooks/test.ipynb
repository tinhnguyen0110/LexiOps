````python
import subprocess
import time

def check_kubernetes_mcp():
    """Check if MCP server is running in Kubernetes"""
    try:
        print("🔍 Checking MCP server in Kubernetes...")
        result = subprocess.run([
            'kubectl', 'get', 'pods', '-n', 'lexiops-copilot', 
            '-l', 'app=mcp-k8s-server'
        ], capture_output=True, text=True, timeout=10)
        
        if result.returncode == 0:
            print("✅ Kubernetes cluster accessible")
            print(f"📋 MCP Pods:\n{result.stdout}")
            
            # Check if pods are running
            if "Running" in result.stdout:
                print("✅ MCP server pods are running")
                return True
            else:
                print("❌ MCP server pods not running")
                return False
        else:
            print(f"❌ kubectl error: {result.stderr}")
            return False
            
    except subprocess.TimeoutExpired:
        print("⏰ kubectl command timed out")
        return False
    except Exception as e:
        print(f"❌ Error checking Kubernetes: {e}")
        return False

def setup_port_forward():
    """Setup port forwarding for MCP server"""
    try:
        print("🔗 Setting up port forwarding...")
        print("💡 Run this command in a separate terminal:")
        print("kubectl port-forward -n lexiops-copilot service/mcp-k8s-server 8002:3000")
        print("\n⚠️ Keep that terminal running while testing!")
        return True
    except Exception as e:
        print(f"❌ Error: {e}")
        return False

# Check Kubernetes status
k8s_ok = check_kubernetes_mcp()

if k8s_ok:
    print("\n" + "="*50)
    setup_port_forward()
else:
    print("\n💡 Alternative: Try running MCP server locally:")
    print("python mcp_servers/utils_server.py")
````

### 2. Simple Chat Test
Test the full 3-tier architecture with a basic request.


````python
def test_fastapi_chat(message):
    """Test FastAPI chat endpoint (Tier 1 → Tier 2 → Tier 3)"""
    try:
        payload = {"message": message}
        response = requests.post(
            f"{FASTAPI_URL}/chat", 
            json=payload, 
            timeout=30
        )
        
        if response.status_code == 200:
            result = response.json()
            print(f"✅ Chat API Success!")
            print(f"📝 Response: {result['response'][:200]}...")
            if result.get('tool_results'):
                print(f"🔧 Tools used: {len(result['tool_results'])}")
            return result
        else:
            print(f"❌ Chat API failed: {response.status_code}")
            print(f"Error: {response.text}")
            return None
    except Exception as e:
        print(f"❌ Chat API error: {e}")
        return None
````

## Troubleshooting Guide

### Common Issues:

1. **FastAPI Connection Failed**: Check if the server is running on `http://localhost:8000`
2. **MCP Tools Not Available**: Verify Kubernetes deployment is running and accessible
3. **Chat API Timeout**: Some Kubernetes operations take longer - increase timeout if needed
4. **Session Errors**: MCP requires proper session initialization - restart if issues persist

### Manual Testing Commands:

```bash
# Check FastAPI health
curl http://localhost:8000/health

# Check Kubernetes deployment
kubectl get pods -n lexiops-copilot

# Port forward MCP service
```
